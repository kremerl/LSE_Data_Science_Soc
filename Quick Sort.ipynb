```python
with open('data_pickled.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
#filter df according to city and cuisine
def filter_cities(city,cuisine):
    data2 = data.loc[data['city'] == city]
    x = data2.loc[data2['Cuisine'] == cuisine]
    return x
    

    
```

```python
#sort filtered df according to stars then review_count
def quick_sort(x):
    x_sorted = x.sort_values(by =['stars','review_count'],axis=0, ascending=False)
    #x_sorted_final =x_sorted.sort_values(by='review_count', axis=0, ascending = False)
    return x_sorted
```

```python
location = input("What city are you looking for? \n")
type_of_food = input("What cuisine are you looking for? \n")

df = filter_cities(location,type_of_food)
quick_sort(df)
```
